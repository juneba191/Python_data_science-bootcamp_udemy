
# Figure size and DPI

```python
import numpy as np
x = np.array([1,2,3,4,4,5,6,6])
y = x** 2
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my_picture.pdf',dpi=200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')


ax.set_title('title')
ax.set_ylabel('Y')
ax.set_xlabel('x')
ax.legend()

```

```python
fig = plt.figure()
ad = fig.add_axes([0,0,1,1])
ad.plot(x,y,color='#FF8C00')
```
