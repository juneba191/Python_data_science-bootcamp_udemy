# Matplotlib overview lecture


```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y  = x**2
```

```python
x
```

```python
y
```

```python
# Functional
plt.plot(x,y)
plt.xlabel('X  Label')
plt.ylabel('Y label')
plt.title('title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# OO
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set title')
```

```python
#create a figure that is a canvas
fig = plt.figure()

#plot the axis on the canvas
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes3 = fig.add_axes([0.4,0.2,0.25,0.25])

#plot the points in the axis.
axes1.plot(x,y)
axes2.plot(y,x)
axes3.plot(y,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes.plot(x,y)
```
